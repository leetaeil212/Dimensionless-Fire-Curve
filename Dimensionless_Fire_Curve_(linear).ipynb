```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
from google.colab import files
import io
import seaborn as sns

# 🔹 모델 클래스 정의 (감쇠구간 포함)
class FireCurveLinearSteady:
    def __init__(self, t_ch, Q_peak, Q_tot=None, t0=0.0):
        self.t_ch = t_ch
        self.Q_peak = Q_peak
        self.Q_tot = Q_tot
        self.beta_i = None
        self.beta_s = None
        self.delta_l = None  # NEW: 감쇠계수 저장
        self.t0 = t0

    def hrr(self, t, beta_i, beta_s, delta_l=None):
        """열방출률 함수 (감쇠구간 포함)"""
        t0 = self.t0
        t_ch = self.t_ch
        Q_peak = self.Q_peak

        t1 = t0 + beta_i * t_ch
        t2 = t0 + (beta_i + beta_s) * t_ch

        if t < t0:
            return 0
        elif t < t1:  # 성장구간 (Linear)
            return Q_peak * (t - t0) / (beta_i * t_ch)
        elif t < t2:  # 유지구간
            return Q_peak
        elif delta_l is not None:  # 감쇠구간 (Linear)
            t_f_star = t2 + (beta_i * t_ch) / delta_l
            if t < t_f_star:
                return Q_peak * (1 - delta_l * (t - t2) / (beta_i * t_ch))
            else:
                return 0
        else:
            return 0

    def calculate_delta_l(self, beta_i, beta_s, Q_tot_exp):
        """
        논문 공식에 따라 감쇠계수 delta_l 계산
        delta_l = β_i / (2 * (Q/(Q_max*Δt_ch) - β_i/2 - β_s))
        """
        Q_normalized = Q_tot_exp / (self.Q_peak * self.t_ch)
        denominator = Q_normalized - (beta_i / 2) - beta_s

        if denominator <= 0:
            raise ValueError("감쇠계수 계산 불가: 분모가 0 이하입니다.")

        delta_l = beta_i / (2 * denominator)
        return delta_l

    def growth_error(self, beta_i, t_exp, Q_exp, alpha=3.0, gamma=2.0):
        t0 = self.t0
        t1 = t0 + beta_i * self.t_ch
        mask = (t_exp >= t0) & (t_exp < t1)
        if np.sum(mask) == 0:
            return float('inf')
        t_growth = t_exp[mask]
        Q_growth_exp = Q_exp[mask]
        Q_growth_model = np.array([self.hrr(t, beta_i, 0) for t in t_growth])
        Q_peak = np.max(Q_growth_exp)
        w = 1 + alpha * (Q_growth_exp / Q_peak) ** gamma
        error = np.sqrt(np.sum(w * (Q_growth_exp - Q_growth_model) ** 2) / np.sum(w))
        return error

    def optimize_parameters(self, t_exp, Q_exp, beta_i_min=0.0, beta_i_max=0.5,
                            alpha=2.5, gamma=2.0, smoothing_window=5, decline_threshold=0.05, plot=True):

        # 1️⃣ 실험 총열발생량 계산
        Q_tot_exp = np.trapz(Q_exp, t_exp)
        print(f"📊 실험 총열발생량: {Q_tot_exp:.0f} kJ")

        # 2️⃣ β_i 최적화 (기존과 동일)
        result_i = minimize_scalar(
            lambda bi: self.growth_error(bi, t_exp, Q_exp, alpha, gamma),
            bounds=(beta_i_min, beta_i_max),
            method='bounded'
        )
        beta_i_opt = result_i.x
        t1 = self.t0 + beta_i_opt * self.t_ch

        # 3️⃣ β_s 계산 (기존 알고리즘과 동일)
        window = min(smoothing_window, len(Q_exp) // 10)
        if window < 3:
            window = 3
        Q_smooth = np.convolve(Q_exp, np.ones(window)/window, mode='valid')
        t_smooth = t_exp[window//2:-(window//2) if window//2 > 0 else None]
        if len(t_smooth) != len(Q_smooth):
            min_len = min(len(t_smooth), len(Q_smooth))
            t_smooth = t_smooth[:min_len]
            Q_smooth = Q_smooth[:min_len]

        # 감쇠 시작점 탐지
        peak_idx = np.argmax(Q_smooth)
        Q_peak_smooth = Q_smooth[peak_idx]
        t1_idx_smooth = np.argmin(np.abs(t_smooth - t1))
        start_idx = max(t1_idx_smooth, peak_idx)
        decline_start_idx = None

        for i in range(start_idx, len(Q_smooth) - window):
            slope = (Q_smooth[i+window] - Q_smooth[i]) / (t_smooth[i+window] - t_smooth[i])
            if slope < -decline_threshold * Q_peak_smooth / (t_smooth[-1] - t_smooth[0]):
                consistent_decline = True
                for j in range(i, min(i + window, len(Q_smooth) - 1)):
                    if Q_smooth[j+1] > Q_smooth[j] * 1.01:
                        consistent_decline = False
                        break
                if consistent_decline:
                    decline_start_idx = i
                    break

        if decline_start_idx is None or decline_start_idx >= len(t_smooth):
            decline_start_idx = int(len(t_smooth) * 0.75)

        decline_start_idx_orig = np.argmin(np.abs(t_exp - t_smooth[decline_start_idx]))
        t_decline_start = t_exp[decline_start_idx_orig]
        beta_s_opt = (t_decline_start - t1) / self.t_ch

        # 4️⃣ NEW: 감쇠계수 계산
        try:
            delta_l_opt = self.calculate_delta_l(beta_i_opt, beta_s_opt, Q_tot_exp)
            print(f"✅ 감쇠계수 계산 성공: δ_l = {delta_l_opt:.4f}")
        except ValueError as e:
            print(f"❌ 감쇠계수 계산 실패: {e}")
            delta_l_opt = None

        # 5️⃣ 최적 비율 저장
        self.beta_i = beta_i_opt
        self.beta_s = beta_s_opt
        self.delta_l = delta_l_opt  # NEW: 감쇠계수 저장

        # 6️⃣ 전체 모델 곡선 생성 (감쇠구간 포함)
        if delta_l_opt is not None:
            t2 = self.t0 + (beta_i_opt + beta_s_opt) * self.t_ch
            t_f_star = t2 + (beta_i_opt * self.t_ch) / delta_l_opt
            t_model = np.linspace(self.t0, t_f_star, 1000)
            Q_model = np.array([self.hrr(t, beta_i_opt, beta_s_opt, delta_l_opt) for t in t_model])
            Q_tot_model = np.trapz(Q_model, t_model)

            print(f"📊 모델 검증:")
            print(f"   실험 총열발생량: {Q_tot_exp:.0f} kJ")
            print(f"   모델 총열발생량: {Q_tot_model:.0f} kJ")
            print(f"   오차: {abs(Q_tot_exp - Q_tot_model)/Q_tot_exp*100:.2f}%")

            # β_d 계산 및 출력
            beta_d_actual = (t_f_star - t2) / self.t_ch
            print(f"📏 구간비:")
            print(f"   β_i = {beta_i_opt:.4f}")
            print(f"   β_s = {beta_s_opt:.4f}")
            print(f"   β_d = {beta_d_actual:.4f} (실제 감쇠구간비)")
            print(f"   합계 = {beta_i_opt + beta_s_opt + beta_d_actual:.4f}")
        else:
            # 감쇠계수 계산 실패시 기존 방식
            t_end = self.t0 + (beta_i_opt + beta_s_opt) * self.t_ch
            t_model = np.linspace(self.t0, t_end, 1000)
            Q_model = np.array([self.hrr(t, beta_i_opt, beta_s_opt) for t in t_model])
            Q_tot_model = np.trapz(Q_model, t_model)

        if plot:
            plt.figure(figsize=(12, 6))
            plt.plot(t_exp, Q_exp, 'o', alpha=0.3, label='Raw Data')
            plt.plot(t_smooth, Q_smooth, '-', linewidth=2, label='Smoothed Data')
            plt.axvline(t1, color='r', linestyle='--', label='Growth End (Optimized)')
            plt.axvline(t_decline_start, color='g', linestyle='--', label='Steady End / Decay Start')
            if delta_l_opt is not None:
                t_f_star = t2 + (beta_i_opt * self.t_ch) / delta_l_opt
                plt.axvline(t_f_star, color='orange', linestyle='--', label='Model End (t_f*)')
            plt.xlabel('Time (s)')
            plt.ylabel('Heat Release Rate (kW)')
            plt.title('Fire Stages Identification (Complete 3S Linear Model)')
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.show()

        # ✅ 🔸 RMSE 계산 (각 구간별)
        growth_rmse = 0
        steady_rmse = 0
        decay_rmse = 0

        if delta_l_opt is not None:
            # 성장구간 RMSE
            mask_growth = (t_exp >= self.t0) & (t_exp < t1)
            if np.sum(mask_growth) > 0:
                Q_exp_growth = Q_exp[mask_growth]
                Q_model_growth = np.array([self.hrr(t, beta_i_opt, beta_s_opt, delta_l_opt) for t in t_exp[mask_growth]])
                growth_rmse = np.sqrt(np.mean((Q_exp_growth - Q_model_growth) ** 2))

            # 유지구간 RMSE
            mask_steady = (t_exp >= t1) & (t_exp < t_decline_start)
            if np.sum(mask_steady) > 0:
                Q_exp_steady = Q_exp[mask_steady]
                Q_model_steady = np.array([self.hrr(t, beta_i_opt, beta_s_opt, delta_l_opt) for t in t_exp[mask_steady]])
                steady_rmse = np.sqrt(np.mean((Q_exp_steady - Q_model_steady) ** 2))

            # 감쇠구간 RMSE
            t_f_star = t2 + (beta_i_opt * self.t_ch) / delta_l_opt
            mask_decay = (t_exp >= t_decline_start) & (t_exp < t_f_star)
            if np.sum(mask_decay) > 0:
                Q_exp_decay = Q_exp[mask_decay]
                Q_model_decay = np.array([self.hrr(t, beta_i_opt, beta_s_opt, delta_l_opt) for t in t_exp[mask_decay]])
                decay_rmse = np.sqrt(np.mean((Q_exp_decay - Q_model_decay) ** 2))

        # ✅ 🔸 Excel 저장 (감쇠구간 포함)
        if delta_l_opt is not None:
            df_curve = pd.DataFrame({
                "Time (s)": t_model,
                "HRR (kW)": Q_model
            })

            df_summary = pd.DataFrame({
                "t_ch (s)": [self.t_ch],
                "alpha*": [alpha],
                "gamma*": [gamma],
                "beta_i": [beta_i_opt],
                "t1 (s)": [t1],
                "t2 (s)": [t_decline_start],
                "beta_s": [beta_s_opt],
                "delta_l": [delta_l_opt],
                "Q_tot_exp (kJ)": [Q_tot_exp],
                "Q_tot_model (kJ)": [Q_tot_model],
                "Error (%)": [abs(Q_tot_exp - Q_tot_model)/Q_tot_exp*100],
                "Growth_RMSE (kW)": [growth_rmse],
                "Steady_RMSE (kW)": [steady_rmse],
                "Decay_RMSE (kW)": [decay_rmse]
            })

            with pd.ExcelWriter("fire_model_prediction_linear.xlsx", engine='openpyxl') as writer:
                df_curve.to_excel(writer, sheet_name="HRR Curve", index=False)
                df_summary.to_excel(writer, sheet_name="Summary", index=False)

            print("💾 결과가 'fire_model_prediction_linear.xlsx'에 저장되었습니다.")
            print(f"📊 구간별 RMSE:")
            print(f"   성장구간: {growth_rmse:.2f} kW")
            print(f"   유지구간: {steady_rmse:.2f} kW")
            print(f"   감쇠구간: {decay_rmse:.2f} kW")

        return beta_i_opt, beta_s_opt

    def plot_comparison(self, t_exp, Q_exp):
        """완전한 3S 모델 (감쇠구간 포함) 비교 그래프"""
        if self.beta_i is None or self.beta_s is None:
            print("먼저 optimize_parameters()를 실행하세요.")
            return

        t0 = self.t0
        t_ch = self.t_ch
        t1 = t0 + self.beta_i * t_ch
        t2 = t0 + (self.beta_i + self.beta_s) * t_ch

        if self.delta_l is not None:
            # 감쇠구간 포함
            t_f_star = t2 + (self.beta_i * t_ch) / self.delta_l
            t_model = np.linspace(self.t0, t_f_star, 1000)
            Q_model = np.array([self.hrr(t, self.beta_i, self.beta_s, self.delta_l) for t in t_model])
            model_name = "Complete 3S Linear Fire Curve (Growth-Steady-Decay)"
        else:
            # 감쇠구간 없음
            t_model = np.linspace(self.t0, t2, 1000)
            Q_model = np.array([self.hrr(t, self.beta_i, self.beta_s) for t in t_model])
            model_name = "Partial Linear Fire Curve (Growth-Steady Only)"

        plt.figure(figsize=(14, 8))
        plt.plot(t_exp, Q_exp, 'o', alpha=0.5, markersize=3, label='Experimental Data')
        plt.plot(t_model, Q_model, '-', linewidth=3, label=model_name)
        plt.axvline(t1, color='r', linestyle='--', alpha=0.7, label='End of Growth')
        plt.axvline(t2, color='g', linestyle='--', alpha=0.7, label='End of Steady')

        if self.delta_l is not None:
            t_f_star = t2 + (self.beta_i * self.t_ch) / self.delta_l
            plt.axvline(t_f_star, color='orange', linestyle='--', alpha=0.7, label='Model End (t_f*)')

        plt.xlabel('Time (s)')
        plt.ylabel('Heat Release Rate (kW)')
        plt.title(f'Complete 3S Linear Fire Curve vs Experimental Data')
        plt.legend()
        plt.grid(True, alpha=0.3)

        # 구간 표시
        plt.text((t0+t1)/2, self.Q_peak*0.3, 'Growth\n(Linear)', ha='center', fontsize=10,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7))
        plt.text(t1 + (t2-t1)/2, self.Q_peak*1.05, 'Steady', ha='center', fontsize=10,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7))

        if self.delta_l is not None:
            t_f_star = t2 + (self.beta_i * self.t_ch) / self.delta_l
            plt.text(t2 + (t_f_star-t2)/2, self.Q_peak*0.5, 'Decay\n(Linear)', ha='center', fontsize=10,
                    bbox=dict(boxstyle="round,pad=0.3", facecolor="lightyellow", alpha=0.7))

        plt.tight_layout()
        plt.show()

# 🔹 다중 재료 처리 함수 (개별 시간축 적용)
def process_multiple_materials():
    """여러 재료를 배치 처리하는 함수 (각 재료별 개별 시간축)"""
    print("🔥 다중 재료 배치 처리를 시작합니다! (Linear Model)")
    print("=" * 60)
    print("📁 여러 Excel 파일을 선택하세요. (재료별로 파일명을 구분해주세요)")
    print("⚠️  파일 선택 창에서 취소하려면 'Cancel' 버튼을 클릭하세요.")

    try:
        uploaded = files.upload()
    except KeyboardInterrupt:
        print("❌ 파일 업로드가 취소되었습니다.")
        return None, None
    except Exception as e:
        print(f"❌ 파일 업로드 중 오류 발생: {e}")
        return None, None

    if len(uploaded) == 0:
        print("❌ 업로드된 파일이 없습니다.")
        return None, None

    alpha_range = np.round(np.arange(1.0, 5.0 + 0.01, 0.1), 2)
    gamma_range = np.round(np.arange(1.0, 5.0 + 0.01, 0.1), 2)

    all_results = []  # 모든 재료의 결과를 저장할 리스트
    all_curves = {}   # 각 재료별 곡선 데이터 (딕셔너리로 변경)

    for filename in uploaded.keys():
        print(f"\n🔍 {filename} 처리 중...")
        try:
            # 데이터 로드
            df = pd.read_excel(io.BytesIO(uploaded[filename]))
            t_exp = df.iloc[:, 0].values
            Q_exp = df.iloc[:, 1].values
            valid = ~np.isnan(t_exp) & ~np.isnan(Q_exp) & (Q_exp >= 0)
            t_exp = t_exp[valid]
            Q_exp = Q_exp[valid]
            if t_exp[0] != 0:
                t_exp = t_exp - t_exp[0]

            t_ch = t_exp[-1]
            Q_peak = np.max(Q_exp)
            Q_tot = np.trapz(Q_exp, t_exp)

            material_name = filename.replace('.xlsx', '').replace('.xls', '')

            print(f"   📊 기본 정보: Q_peak = {Q_peak:.0f} kW, t_ch = {t_ch:.0f} s")

            # 최적 alpha, gamma 탐색 (출력 없이)
            print("   🔍 최적 파라미터 탐색 중...")
            best_rmse = float('inf')
            best_alpha = None
            best_gamma = None

            for alpha in alpha_range:
                for gamma in gamma_range:
                    model_temp = FireCurveLinearSteady(t_ch, Q_peak)
                    try:
                        beta_i, beta_s = model_temp.optimize_parameters(
                            t_exp, Q_exp, alpha=alpha, gamma=gamma, plot=False
                        )

                        # 성장구간 RMSE 계산
                        t0 = model_temp.t0
                        t1 = t0 + beta_i * t_ch
                        mask = (t_exp >= t0) & (t_exp < t1)
                        if np.sum(mask) > 0:
                            Q_exp_growth = Q_exp[mask]
                            Q_model_growth = np.array([model_temp.hrr(t, beta_i, 0) for t in t_exp[mask]])
                            rmse = np.sqrt(np.mean((Q_exp_growth - Q_model_growth) ** 2))

                            if rmse < best_rmse:
                                best_rmse = rmse
                                best_alpha = alpha
                                best_gamma = gamma
                    except:
                        continue

            if best_alpha is None:
                print(f"   ❌ {filename}: 최적화 실패")
                continue

            print(f"   ✅ 최적값: α={best_alpha}, γ={best_gamma}, RMSE={best_rmse:.2f}")

            # 최종 모델 생성 및 최적화
            model = FireCurveLinearSteady(t_ch, Q_peak, Q_tot)
            model.optimize_parameters(t_exp, Q_exp, alpha=best_alpha, gamma=best_gamma, plot=False)

            # 결과 수집
            if model.delta_l is not None:
                # 각 구간별 RMSE 계산
                t1 = model.t0 + model.beta_i * model.t_ch
                t2 = model.t0 + (model.beta_i + model.beta_s) * model.t_ch
                t_f_star = t2 + (model.beta_i * model.t_ch) / model.delta_l

                # 실험 총열발생량
                Q_tot_exp = np.trapz(Q_exp, t_exp)

                # 모델 총열발생량
                t_model = np.linspace(model.t0, t_f_star, 1000)
                Q_model = np.array([model.hrr(t, model.beta_i, model.beta_s, model.delta_l) for t in t_model])
                Q_tot_model = np.trapz(Q_model, t_model)

                # 구간별 RMSE
                mask_growth = (t_exp >= model.t0) & (t_exp < t1)
                mask_steady = (t_exp >= t1) & (t_exp < t2)
                mask_decay = (t_exp >= t2) & (t_exp < t_f_star)

                growth_rmse = 0
                steady_rmse = 0
                decay_rmse = 0

                if np.sum(mask_growth) > 0:
                    Q_exp_growth = Q_exp[mask_growth]
                    Q_model_growth = np.array([model.hrr(t, model.beta_i, model.beta_s, model.delta_l) for t in t_exp[mask_growth]])
                    growth_rmse = np.sqrt(np.mean((Q_exp_growth - Q_model_growth) ** 2))

                if np.sum(mask_steady) > 0:
                    Q_exp_steady = Q_exp[mask_steady]
                    Q_model_steady = np.array([model.hrr(t, model.beta_i, model.beta_s, model.delta_l) for t in t_exp[mask_steady]])
                    steady_rmse = np.sqrt(np.mean((Q_exp_steady - Q_model_steady) ** 2))

                if np.sum(mask_decay) > 0:
                    Q_exp_decay = Q_exp[mask_decay]
                    Q_model_decay = np.array([model.hrr(t, model.beta_i, model.beta_s, model.delta_l) for t in t_exp[mask_decay]])
                    decay_rmse = np.sqrt(np.mean((Q_exp_decay - Q_model_decay) ** 2))

                # 결과 저장
                result = {
                    "Material": material_name,
                    "t_ch (s)": model.t_ch,
                    "alpha*": best_alpha,
                    "gamma*": best_gamma,
                    "beta_i": model.beta_i,
                    "t1 (s)": t1,
                    "t2 (s)": t2,
                    "beta_s": model.beta_s,
                    "delta_l": model.delta_l,
                    "Q_tot_exp (kJ)": Q_tot_exp,
                    "Q_tot_model (kJ)": Q_tot_model,
                    "Error (%)": abs(Q_tot_exp - Q_tot_model)/Q_tot_exp*100,
                    "Growth_RMSE (kW)": growth_rmse,
                    "Steady_RMSE (kW)": steady_rmse,
                    "Decay_RMSE (kW)": decay_rmse
                }
                all_results.append(result)

                # 🔹 NEW: 각 재료별 개별 곡선 데이터 저장
                all_curves[material_name] = {
                    "Time (s)": t_model,
                    "HRR (kW)": Q_model
                }

                print(f"   ✅ {material_name} 처리 완료!")
            else:
                print(f"   ❌ {material_name}: 감쇠계수 계산 실패")

        except Exception as e:
            print(f"   ❌ {filename} 처리 실패: {e}")
            continue

    if len(all_results) == 0:
        print("❌ 처리된 재료가 없습니다.")
        return None, None

    # 결과 통합 및 저장
    print(f"\n📊 총 {len(all_results)}개 재료 처리 완료!")

    # 통합 결과 DataFrame
    df_summary_all = pd.DataFrame(all_results)

    # 🔹 NEW: 각 재료별 개별 시트로 곡선 데이터 저장
    with pd.ExcelWriter("multi_material_results_linear.xlsx", engine='openpyxl') as writer:
        # 통합 요약 시트
        df_summary_all.to_excel(writer, sheet_name="Summary_All_Materials", index=False)

        # 각 재료별 개별 곡선 시트
        for material_name, curve_data in all_curves.items():
            df_curve = pd.DataFrame(curve_data)
            sheet_name = f"HRR_{material_name}"[:31]  # Excel 시트명 길이 제한
            df_curve.to_excel(writer, sheet_name=sheet_name, index=False)

    print("💾 통합 결과가 'multi_material_results_linear.xlsx'에 저장되었습니다!")
    print("📋 각 재료별 HRR 곡선이 개별 시트에 저장되었습니다.")

    # 결과 요약 출력
    print("\n📋 재료별 결과 요약:")
    print("=" * 80)
    for result in all_results:
        print(f"{result['Material']:15s} | β_i={result['beta_i']:.3f} | β_s={result['beta_s']:.3f} | δ_l={result['delta_l']:.3f} | Error={result['Error (%)']:.1f}%")

    # 통합 시각화
    plot_multi_material_comparison(all_results, all_curves)

    return df_summary_all, all_curves

# 🔹 다중 재료 비교 시각화 함수 (개별 시간축 적용)
def plot_multi_material_comparison(all_results, all_curves):
    """다중 재료 결과 비교 시각화 (개별 시간축 적용)"""

    # 1. 파라미터 비교 차트
    materials = [r['Material'] for r in all_results]
    beta_i_values = [r['beta_i'] for r in all_results]
    beta_s_values = [r['beta_s'] for r in all_results]
    delta_l_values = [r['delta_l'] for r in all_results]
    errors = [r['Error (%)'] for r in all_results]

    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    # β_i 비교
    axes[0,0].bar(materials, beta_i_values, color='skyblue', alpha=0.7)
    axes[0,0].set_title('Growth Phase Ratio (β_i)')
    axes[0,0].set_ylabel('β_i')
    axes[0,0].tick_params(axis='x', rotation=45)

    # β_s 비교
    axes[0,1].bar(materials, beta_s_values, color='lightgreen', alpha=0.7)
    axes[0,1].set_title('Steady Phase Ratio (β_s)')
    axes[0,1].set_ylabel('β_s')
    axes[0,1].tick_params(axis='x', rotation=45)

    # δ_l 비교
    axes[1,0].bar(materials, delta_l_values, color='lightyellow', alpha=0.7)
    axes[1,0].set_title('Linear Decay Coefficient (δ_l)')
    axes[1,0].set_ylabel('δ_l')
    axes[1,0].tick_params(axis='x', rotation=45)

    # 오차율 비교
    axes[1,1].bar(materials, errors, color='lightcoral', alpha=0.7)
    axes[1,1].set_title('Total Heat Release Error (%)')
    axes[1,1].set_ylabel('Error (%)')
    axes[1,1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.suptitle('Multi-Material Linear Model Parameter Comparison', y=1.02, fontsize=16)
    plt.show()

    # 2. 곡선 비교 차트 (개별 시간축)
    plt.figure(figsize=(14, 8))
    colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']

    for i, (material_name, curve_data) in enumerate(all_curves.items()):
        plt.plot(curve_data["Time (s)"], curve_data["HRR (kW)"],
                linewidth=2.5, label=material_name, color=colors[i % len(colors)])

    plt.xlabel('Time (s)')
    plt.ylabel('Heat Release Rate (kW)')
    plt.title('Multi-Material Linear Fire Curve Comparison (Individual Time Axes)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    # 3. 정규화된 시간축 비교 (0~1 스케일)
    plt.figure(figsize=(14, 8))

    for i, (material_name, curve_data) in enumerate(all_curves.items()):
        # 시간 정규화 (0~1)
        t_normalized = curve_data["Time (s)"] / curve_data["Time (s)"].max()
        # HRR 정규화 (최댓값으로 나누기)
        hrr_normalized = curve_data["HRR (kW)"] / curve_data["HRR (kW)"].max()

        plt.plot(t_normalized, hrr_normalized,
                linewidth=2.5, label=material_name, color=colors[i % len(colors)])

    plt.xlabel('Normalized Time (t/t_f*)')
    plt.ylabel('Normalized HRR (Q̇/Q̇_max)')
    plt.title('Multi-Material Linear Fire Curve Comparison (Normalized)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# 🔹 민감도 자동 탐색 (기존과 동일)
def find_optimal_alpha_gamma(t_exp, Q_exp, t_ch, Q_peak, alpha_range, gamma_range):
    best_rmse = float('inf')
    best_alpha = None
    best_gamma = None
    best_beta_i = None
    best_beta_s = None
    results = []

    for alpha in alpha_range:
        for gamma in gamma_range:
            model = FireCurveLinearSteady(t_ch, Q_peak)
            try:
                beta_i, beta_s = model.optimize_parameters(
                    t_exp, Q_exp,
                    alpha=alpha,
                    gamma=gamma,
                    plot=False
                )
                t0 = model.t0
                t1 = t0 + beta_i * t_ch
                mask = (t_exp >= t0) & (t_exp < t1)
                if np.sum(mask) > 0:
                    Q_exp_growth = Q_exp[mask]
                    Q_model_growth = np.array([model.hrr(t, beta_i, 0) for t in t_exp[mask]])
                    rmse = np.sqrt(np.mean((Q_exp_growth - Q_model_growth) ** 2))
                    rmse = round(rmse, 2)
                    results.append((round(alpha, 2), round(gamma, 2), rmse, round(beta_i, 4), round(beta_s, 4)))

                    if rmse < best_rmse:
                        best_rmse = rmse
                        best_alpha = alpha
                        best_gamma = gamma
                        best_beta_i = beta_i
                        best_beta_s = beta_s
            except:
                continue

    print(f"✔ 자동 탐색된 최적 값 → alpha = {best_alpha}, gamma = {best_gamma}, beta_i = {best_beta_i:.4f}, beta_s = {best_beta_s:.4f}, RMSE = {best_rmse:.2f}")

    # CSV 저장
    df_result = pd.DataFrame(results, columns=["alpha", "gamma", "growth_rmse", "beta_i", "beta_s"])
    df_result.to_csv("sensitivity_results_linear.csv", index=False)
    print("📁 민감도 분석 결과가 'sensitivity_results_linear.csv' 파일로 저장되었습니다.")

    # 히트맵 시각화
    pivot_table = df_result.pivot(index='alpha', columns='gamma', values='growth_rmse')
    plt.figure(figsize=(10, 6))
    sns.heatmap(pivot_table, annot=False, cmap="YlGnBu", cbar_kws={'label': 'Growth RMSE (kW)'})
    plt.title("RMSE Sensitivity to α and γ (Linear Model)")
    plt.xlabel("γ (gamma)")
    plt.ylabel("α (alpha)")
    plt.tight_layout()
    plt.show()

    return best_alpha, best_gamma

# 🔹 데이터 불러오기 (에러 처리 강화)
def load_and_process_data():
    print("📁 엑셀 파일을 업로드하세요.")
    print("⚠️  파일 선택 창에서 취소하려면 'Cancel' 버튼을 클릭하세요.")

    try:
        uploaded = files.upload()
    except KeyboardInterrupt:
        print("❌ 파일 업로드가 취소되었습니다.")
        return None, None, None, None, None
    except Exception as e:
        print(f"❌ 파일 업로드 중 오류 발생: {e}")
        return None, None, None, None, None

    if len(uploaded) == 0:
        print("❌ 업로드된 파일이 없습니다.")
        return None, None, None, None, None

    for filename in uploaded.keys():
        try:
            df = pd.read_excel(io.BytesIO(uploaded[filename]))
            print(f"✅ {filename} 불러오기 성공")
            t_exp = df.iloc[:, 0].values
            Q_exp = df.iloc[:, 1].values
            valid = ~np.isnan(t_exp) & ~np.isnan(Q_exp) & (Q_exp >= 0)
            t_exp = t_exp[valid]
            Q_exp = Q_exp[valid]
            if t_exp[0] != 0:
                t_exp = t_exp - t_exp[0]
            t_ch = t_exp[-1]
            Q_peak = np.max(Q_exp)
            Q_tot = np.trapz(Q_exp, t_exp)
            return t_exp, Q_exp, t_ch, Q_peak, Q_tot
        except Exception as e:
            print(f"❌ {filename} 처리 중 오류 발생: {e}")
            return None, None, None, None, None

# 🔹 실행 부분
print("🔥 3S Linear Fire Curve 분석 툴 (개별 시간축 적용)")
print("=" * 60)
print("1️⃣ 단일 재료 분석")
print("2️⃣ 다중 재료 배치 처리 (개별 시간축)")

# Colab에서는 input() 대신 다른 방식 사용
try:
    choice = input("선택하세요 (1 또는 2): ")
except:
    choice = "1"  # 기본값

if choice == "1":
    # 기존 단일 재료 처리
    alpha_range = np.round(np.arange(1.0, 5.0 + 0.01, 0.1), 2)
    gamma_range = np.round(np.arange(1.0, 5.0 + 0.01, 0.1), 2)

    t_exp, Q_exp, t_ch, Q_peak, Q_tot = load_and_process_data()
    if t_exp is not None:
        opt_alpha, opt_gamma = find_optimal_alpha_gamma(t_exp, Q_exp, t_ch, Q_peak, alpha_range, gamma_range)
        model = FireCurveLinearSteady(t_ch, Q_peak, Q_tot)
        model.optimize_parameters(t_exp, Q_exp, alpha=opt_alpha, gamma=opt_gamma)
        model.plot_comparison(t_exp, Q_exp)

elif choice == "2":
    # 다중 재료 처리 (개별 시간축)
    result = process_multiple_materials()

    if result is not None:
        df_summary, all_curves = result

        # 추가: 각 재료별 파라미터 상세 분석
        if len(df_summary) > 0:
            print("\n📊 재료별 상세 분석:")
            print("=" * 80)

            # 구간비 통계
            print(f"β_i 범위: {df_summary['beta_i'].min():.3f} ~ {df_summary['beta_i'].max():.3f} (평균: {df_summary['beta_i'].mean():.3f})")
            print(f"β_s 범위: {df_summary['beta_s'].min():.3f} ~ {df_summary['beta_s'].max():.3f} (평균: {df_summary['beta_s'].mean():.3f})")
            print(f"δ_l 범위: {df_summary['delta_l'].min():.3f} ~ {df_summary['delta_l'].max():.3f} (평균: {df_summary['delta_l'].mean():.3f})")
            print(f"오차율 범위: {df_summary['Error (%)'].min():.1f}% ~ {df_summary['Error (%)'].max():.1f}% (평균: {df_summary['Error (%)'].mean():.1f}%)")
    else:
        print("🔄 다중 재료 처리가 취소되었습니다. 단일 재료 분석으로 전환합니다.")
        # 단일 재료 분석으로 전환
        alpha_range = np.round(np.arange(1.0, 5.0 + 0.01, 0.1), 2)
        gamma_range = np.round(np.arange(1.0, 5.0 + 0.01, 0.1), 2)

        t_exp, Q_exp, t_ch, Q_peak, Q_tot = load_and_process_data()
        if t_exp is not None:
            opt_alpha, opt_gamma = find_optimal_alpha_gamma(t_exp, Q_exp, t_ch, Q_peak, alpha_range, gamma_range)
            model = FireCurveLinearSteady(t_ch, Q_peak, Q_tot)
            model.optimize_parameters(t_exp, Q_exp, alpha=opt_alpha, gamma=opt_gamma)
            model.plot_comparison(t_exp, Q_exp)

else:
    print("❌ 잘못된 선택입니다. 단일 재료 분석을 실행합니다.")
    # 기본값으로 단일 재료 처리
    alpha_range = np.round(np.arange(1.0, 5.0 + 0.01, 0.1), 2)
    gamma_range = np.round(np.arange(1.0, 5.0 + 0.01, 0.1), 2)

    t_exp, Q_exp, t_ch, Q_peak, Q_tot = load_and_process_data()
    if t_exp is not None:
        opt_alpha, opt_gamma = find_optimal_alpha_gamma(t_exp, Q_exp, t_ch, Q_peak, alpha_range, gamma_range)
        model = FireCurveLinearSteady(t_ch, Q_peak, Q_tot)
        model.optimize_parameters(t_exp, Q_exp, alpha=opt_alpha, gamma=opt_gamma)
        model.plot_comparison(t_exp, Q_exp)
```
